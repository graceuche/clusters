# K-Means Clustering Example

[Cluster analysis or clustering](https://en.wikipedia.org/wiki/Cluster_analysis) is the task of grouping a set of objects in such a way that objects in the same group (called a cluster) are more similar (in some sense) to each other than to those in other groups (clusters).

Let's make some fake data that includes people clustered by income and age, randomly:

```python
import numpy as np # imports numpy package

#Create fake income/age clusters
def create_clustered_data(num_people, clust): # creates a function create_clustered_data
    np.random.seed(10)
    points_per_cluster = num_people/clust # calculates the number of people per cluster
    clustered_data = [] # initializes clustered_data list
    for i in range (clust): # loops clust number of times e.g 5 or so 
        income_centroid = np.random.uniform(20000.0, 200000.0) # creates a random float between 20,000 - 200,000 
        age_centroid = np.random.uniform(20.0, 70.0) # creates a random float between 20-70 
        for j in range(int(points_per_cluster)): # loops as many times as we want people per cluster
            clustered_data.append([np.random.normal(income_centroid, 10000.0), np.random.normal(age_centroid, 2.0)]) # gets random number not far from the centroid of a cluster
    clustered_data = np.array(clustered_data) # converts list into numpy array
    return clustered_data # returns numpy array of clust number of clusters
```

We'll use k-means to rediscover these clusters in unsupervised learning:

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random

data = create_clustered_data(100, 6)

model = KMeans(n_clusters=6)

# Note I'm scaling the data to normalize it! Important for good results.
model = model.fit(scale(data))

# We can look at the clusters each data point was assigned to
# print(model.labels_)

# And we'll visualize it:
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```
